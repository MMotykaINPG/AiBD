```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("analysis_data.csv")
df
```

```python
print('Kolumna religion zawiera religię, którą zadeklarował badany  \n') 

print(df['religion'].describe())
print('\n')
print(df['religion'].value_counts())

print(' \n W pliku występuje 65 różnych religii, każda z nich została ma zbadanych 10 przedziałów dochodów.')
```

```python
print('Kolumna income zawiera przedziały dochodów, które zostały zawarte w badaniu \n') 

print(df['income'].describe())
print('\n')
print(df['income'].value_counts())

print(' \n W badaniu przewidziano 10 różnych przedziałów, z czego jeden oznacza "Nie wiem/nie chce odpowiadać". ')
```

```python
print('Ostatnia kolumna - freq - opisuje jak często wyznawca danej religii deklarował dany przychód')

print(df['freq'].describe())
print('\n')


df.head()
```
